```python
import cv2
import csv
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from utils import preprocess_image
```

```python
BASE_CSV_PATH = u'data'
BASE_IMG_PATH = u'data/img'
BIAS = 10.0
IMAGE_DIM = (320, 70)  # (240, 135)
```

```python
with open(u'{}/training_data.csv'.format(BASE_CSV_PATH)) as csvfile:
    lines = csvfile.readlines()[2190:2191]
    reader = csv.reader(lines, delimiter=',')
    for line in reader:
        img_id = line[0]
        img_path = u'{}/center-{}.jpg'.format(BASE_IMG_PATH, img_id)
        img_orig = mpimg.imread(img_path)
        image = preprocess_image(cv2.imread(img_path), crop=True)
        img_final = image
```
